Q2. What happens if stopwords are not removed?

A. If we don’t remove stopwords like “is”, “the”, “and”, the data can become a bit messy and unnecessarily large. These words don’t add much meaning, so they can affect model performance. But at the same time, removing them blindly is not always a good idea.

Q3. Provide two real-world scenarios where removing stopwords can be harmful.
A. 	
In sentiment analysis, words like “not” are very important.
For example, “I am not happy” becomes “I am happy” if we remove “not”, which completely changes the meaning.
In question-based systems, words like “what”, “where”, “when” are important.
If we remove them, the sentence might not make sense or the model may not understand the question properly.
