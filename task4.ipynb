```python
import re, numpy as np
def preprocess_text(text):
    try:
        if not text or not isinstance(text, str):
            return []
        text = text.lower()
        text = re.sub(r'http\S+|www\S+', '', text)
        text = re.sub(r'\S+@\S+', '', text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'(.)\1{2,}', r'\1', text)
        text = re.sub(r'[^a-z\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        tokens = text.split()
        cleaned_tokens = []
        for word in tokens:
            if len(word) > 2 or word in ["no", "not"]:
                cleaned_tokens.append(word)
        return cleaned_tokens
    except Exception as e:
        print("error:", e)
        return []
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://instagram.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
analysis_results = []

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = np.mean([len(word) for word in tokens]) if tokens else 0

    analysis_results.append({
        "sentence": sentence,
        "tokens": tokens,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": round(avg_token_length, 2)
    })

    print(f"\n--- Sentence {i} ---")
    print("Total Tokens      :", total_tokens)
    print("Unique Tokens     :", unique_tokens)
    print("Avg Token Length  :", round(avg_token_length, 2))
```
